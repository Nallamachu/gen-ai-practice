#### Text file example using python

These are the different ways of creating the file in current directory.

##### Using OS module

```python
import os
current_directory = os.getcwd()

# Print the current directory
print("Current directory:", current_directory)

file_path = os.path.join(current_directory, "file.txt")

with open(file_path, "w") as file:
    file.write("Hello, this is a test file.")
    
print(f"File created at: {file_path}")
```

##### Using pathlib

```python
from pathlib import Path

# Create a Path object for the file
current_path = file_path = Path.cwd()

file_path = current_path / "sample_file.txt"

file_path.write_text("This is a sample file created using pathlib.")

print(f"File created at: {file_path}")
```

##### Read all text files in current path and print the text

```python
# OS already imported in previous examples hence not importing again
import os
current_path = os.getcwd()

# Loop through all files in the current directory
for filename in os.listdir(current_path):
    if(filename.endswith(".txt")):
        file_path = os.path.join(current_path, filename)
    
        with open(file_path, "r", encoding="utf-8") as file:
            content = file.read()
            print(f"Content of {filename}: {content}")
```

##### Append content in existing file

```python
import os
current_path = os.getcwd()
file = os.path.join(current_path, "file.txt")

lines = ["Line 1", "Line 2", "Line 3"]

with open(file, "a") as file:
    file.write("\n".join(lines))
    
print(f"Lines appended to {file}")
```
