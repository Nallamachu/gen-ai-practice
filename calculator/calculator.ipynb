#### Calculator

```python
def calculate(a, b):
    try:
        result = a / b
    except ZeroDivisionError as ex:
        print(ex)
        print("You cannot divide by zero. Please provide a non-zero denominator.")
    except NameError as ex:
        print(ex)
        print("The variable 'b' is not defined. Please define it before using.")
    else:
        print(f"The result is {result}")
    finally:
        print("Execution completed. Finally block runs regardless of exceptions.")
        
calculate(1, 0)
calculate(1, 2)
```

##### We will look into the calculate operation by using the match/switch statement

```python
def calculate(a, b, operation):
    try:
        match operation:
            case 'addition':
                result = a + b
            case 'subtraction':
                result = a - b
            case 'divide':
                result = a / b
            case 'multiply':
                result = a * b
            case _:
                raise ValueError("Unsupported operation. Please use 'divide' or 'multiply'.")
    except ValueError as ex:
        print(ex)
    else:
        print(f"The result is {result}")
    finally:
        print("Execution completed. Finally block runs regardless of exceptions.")
        
calculate(1, 2, 'divide')
calculate(1, 2, 'multiply')
calculate(1, 2, 'addition')
calculate(1, 2, 'subtraction')
```

##### We will write the same logic with if-else statements

```python
def calculate(a, b, operation):
    try:
        if operation == 'divide':
            result = a / b
        elif operation == 'multiply':
            result = a * b
        elif operation == 'addition':
            result = a + b
        elif operation == 'subtraction':
            result = a - b
        else:
            raise ValueError("Unsupported operation. Please use 'divide' or 'multiply'.")
    except ZeroDivisionError as ex:
        print(ex)
        print("You cannot divide by zero. Please provide a non-zero denominator.")
    except NameError as ex:
        print(ex)
        print("The variable 'b' is not defined. Please define it before using.")
    except ValueError as ex:
        print(ex)
    else:
        print(f"The result of {operation} is {result}")
    finally:
        print("Execution completed. Finally block runs regardless of exceptions.")

calculate(1, 0, 'divide')
calculate(1, 2, 'divide')
calculate(1, 2, 'multiply')
calculate(1, 2, 'addition')
calculate(1, 2, 'subtraction')
```
