#### Exception Handling

```python
a=b
```

##### Above is one of the exception raising by the system. In the below lines you can see how to handle that exception using try and except blocks.

```python
try:
    a=b
except:
    print("The variable 'b' is not defined. Please define it before using.")
```

##### Here is an another exception handling of ZeroDivisionError using try and except blocks

```python
try:
    result=1/0
except ZeroDivisionError as ex:
    print(ex)
    print("You cannot divide by zero. Please provide a non-zero denominator.")
```

##### Here is an example of multi exception handling:

```python
try:
    result=1/2
    a=b
except ZeroDivisionError as ex:
    print(ex)
    print("You cannot divide by zero. Please provide a non-zero denominator.")
except NameError as ex:
    print(ex)
    print("The variable 'b' is not defined. Please define it before using.")
```

```python
try:
    number = int(input("Enter a number: "))
    print(f"You entered: {number}")
    result = 10 / number
    print(f"10 divided by {number} is {result}")
except ValueError as ex:
    print(ex)
    print("Invalid input. Please enter a valid integer.")
except ZeroDivisionError as ex:
    print(ex)
    print("You cannot divide by zero. Please provide a non-zero denominator.")
else:
    print(f"the result is {result}")
```

```python
try:
    number = int(input("Enter a number: "))
    print(f"You entered: {number}")
    result = 10 / number
    print(f"10 divided by {number} is {result}")
except ValueError as ex:
    print(ex)
    print("Invalid input. Please enter a valid integer.")
except ZeroDivisionError as ex:
    print(ex)
    print("You cannot divide by zero. Please provide a non-zero denominator.")
else:
    print(f"the result is {result}")
finally:
    print("Execution completed. Finally block runs regardless of exceptions.")
```

```python
try:
    file=open("non_existent_file.txt", "r")
    content=file.read()
    print(content)
except FileNotFoundError as ex:
    print(ex)
    print("The file does not exist. Please provide a valid file path.")
except IOError as ex:
    print(ex)
    print("An error occurred while reading the file. Please check the file permissions and try again.")
finally:
    if 'file' in locals() and not file.closed:
        file.close()
        print("File closed.")
```
