#### Inheritance in Python

```python
class Car:
    def __init__(self, name, model):
        self.name = name
        self.model = model
    
    def display_info(self):
        print(f"Car Name: {self.name}, Model: {self.model}")
        
    def drive(self):
        print(f"The person driving {self.name} - {self.model} car.")
        
bmw = Car("BMW", "X5")
audi = Car("Audi", "A4")

bmw.display_info()
audi.display_info()

bmw.drive()
audi.drive()
```

```python
class Tesla(Car):
    def __init__(self, name, model, self_driving):
        super().__init__(name, model)
        self.self_driving = self_driving
        
    def drive(self):
        print(f"The person driving {self.name} - {self.model} electric car with {self.self_driving} feature.")
        
tesla = Tesla("Tesla", "Model S", "Autopilot")
tesla.drive()
tesla.display_info()
```

```python
class Audi(Car):
    def __init__(self, name, model, luxury_package):
        super().__init__(name, model)
        self.luxury_package = luxury_package
        
    def display_info(self):
        super().display_info()
        print(f"Luxury Package: {self.luxury_package}")

audi_luxury = Audi("Audi", "A6", "Premium Plus")
audi_luxury.display_info()
```

#### Multiple Inheritance

##### When a class inherit from more than one base class

```python
# Base class 1
class Animal:
    def __init__(self, name):
        self.name = name
        
    def speak(self):
        print(f"{self.name} makes a sound.")

# Base class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner
        
    def show_owner(self):
        print(f"This pet belongs to {self.owner}.")
```

```python
# Derived class inheriting from both Animal and Pet
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)
        
    def speak(self):
        print(f"{self.name} barks.")

dog = Dog("Buddy", "Reddy")
dog.speak()
dog.show_owner()
```
