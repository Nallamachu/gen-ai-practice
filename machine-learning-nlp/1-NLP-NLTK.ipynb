- Chatbots

#### What is NLTK?

*Natural Language Toolkit (NLTK)* is a popular Python library used for Natural Language Processing (NLP).
It provides tools and datasets that help computers understand, analyze, and process human language such as English text.

**NLTK is widely used for:**

- Text processing
- Language analysis
- Machine learning for text
- NLP research and education

It works with Python and supports many NLP tasks.

**1. Tokenization**

Breaking text into words or sentences.

```python
from nltk.tokenize import word_tokenize

text = "Python is great for NLP."
tokens = word_tokenize(text)

print(tokens)
```

**2. Stop Word Removal**

Removing common words that don't add much meaning.

Example:
is, the, and, in, on

```python
from nltk.corpus import stopwords
```

**3. Stemming**

Reducing words to their root form.

Example:
- running → run
- playing → play

**4. Lemmatization**

Converting words to their meaningful base form.

Example:
- better → good
- running → run

**5. Part of Speech (POS) Tagging**

Identifying grammar roles of words.

Example:

- Python (Noun)
- run (Verb)
- beautiful (Adjective)

**6. Named Entity Recognition (NER)**

Identifying names of people, places, organizations.

Example:
- "Elon Musk founded Tesla"
- Person → Elon Musk
- Organization → Tesla

2️⃣ **Real-World Example Using NLTK**
Problem

Analyze customer feedback and remove unnecessary words.

Example Text
*"The product quality is very good and I am very happy with the service"*

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Sample text
text = "The product quality is very good and I am very happy with the service"

# Tokenization
tokens = word_tokenize(text)

# Remove stop words
stop_words = set(stopwords.words("english"))
filtered_words = [word for word in tokens if word.lower() not in stop_words]

print("Tokens:", tokens)
print("Filtered Words:", filtered_words)
```

This helps in tasks like:
- Sentiment analysis
- Product review analysis
- Chatbots
- Recommendation systems
