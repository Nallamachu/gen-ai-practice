#### Stemming

Stemming is the process of reducing a word to it's word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. Stemming is important in natural language understanding(NLU) and natural language processing (NLP).

```python
# Classification problem
# Comments of a product is positive review or negative review
# Reviews ---> eating, eat, eaten [going, goes, gone] ---> eat
# Stemming: eat, goes, gone

words = ["eating", "eats", "eaten","Writing", "writes", "written","programming", "program", "programs", "going", "goes", "gone"]
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
for word in words:
    print(word, "---->", stemmer.stem(word))
```

```python
words = ["eating", "eats", "eaten","Writing", "writes", "written","programming", "program", "programs", "finally", "finalize", "final"]
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")
for word in words:
    print(word, "---->", stemmer.stem(word))
```

```python
stemmer.stem("congratulations")
```

```python
stemmer.stem("sitting")
```


##### RegexpStemmer Class

NLTK has RegexpStemmer class with the help of which we can easily implement regular expression stemmer algorithm. It basically takes a single regular expression and remove prefix or suffix that matches the expression. Let us see an example

```python
# RegexpStemmer("ing") removes ing from the end of the word if it is present. 
# It does not check for the meaning of the word or any other rules. 
# It simply removes "ing" from the end of the word if it is there. 
# So, "congratulations" becomes "congratulations" and "sitting" becomes "sitt".

words = ["eating", "eats", "congratulations","Writing", "writes", "written","programming", "program", "programs", "finally", "finalize", "final"]
from nltk.stem import RegexpStemmer
stemmer = RegexpStemmer("ing")
for word in words:
    print(word, "---->", stemmer.stem(word))
```

```python
words = ["eating", "eats", "congratulations","Writing", "writes", "written","programming", "program", "programs", "finally", "finalize", "final"]
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|ed$|es$|ly$|able$|ible$|ment$|ent$|ous$|ive$|ize$|ise$', min=4)
for word in words:
    print(word, "---->", reg_stemmer.stem(word))
```

#### Snowball Stemmer
